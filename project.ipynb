Import potrzebnych bibliotek

```python
import pandas as pd
import pymssql
from pandas import ExcelWriter
import numpy as np
import xlsxwriter
```

Łączenie z bazą danych

```python
conn = pymssql.connect(server='mssql-2016.labs.wmi.amu.edu.pl', database='dbad_kol_skoki_narciarskie')
```

Wysłanie odpowiednich zapytań do bazy danych za pomocą standardowych zapytań SQL przy pomocy biblioteki pandas

```python
data_1 = pd.read_sql_query("SELECT p.miejsce, z.nazwisko_i_imie, p.wynik_I_serii, p.wynik_II_serii, p.nota, za.nazwa_skoczni, za.data_zawodow FROM Puchar P JOIN Zawodnicy Z ON Z.id_zawodnika = P.zawodnik JOIN Zawody Za ON P.zawody = Za.id_zawodow", conn)
data_2 = pd.read_sql_query("SELECT nazwa_skoczni, sezon, count(*) [ile zawodów] FROM Zawody GROUP BY nazwa_skoczni, sezon", conn)
data_3 = pd.read_sql_query("SELECT panstwo, nazwisko_i_imie, count(*) [ile zawodników] FROM Zawodnicy GROUP BY panstwo, nazwisko_i_imie", conn)
```

Zmiana nazw kolumn

```python
data_1.columns = ['Miejsce', 'Nazwisko i imię', 'Wynik I serii', 'Wynik II serii', 'Nota', 'Nazwa skoczni', 'Data zawodów']
data_2.columns = ['Nazwa skoczni', 'Sezon', 'Liczba zawodów']
data_3.columns = ['Państwo', 'Nazwisko i imię', 'Liczba zawodników']
```

Utworzenie pliku programu Excel, do którego będą zapisywane wygenerowane poszczególne raporty 

```python
writer = ExcelWriter('project.xlsx')
```

Zapisywanie pierwszego raportu do arkusza o nazwie Raport_1

```python
data_1.to_excel(writer, 'Raport_1', index=False)
workbook = writer.book
```

Utworzenie tabeli przestawnej z pierwszego raportu

```python
table_1 = pd.pivot_table(data_1, index=["Nazwisko i imię"], values=["Nota"],
                              aggfunc=[np.sum], fill_value=0)
table_1
```

Zapis pierwszej tabeli przestawnej do arkusza o nazwie Tabela_1

```python
table_1.to_excel(writer, 'Tabela_1')
workbook = writer.book
```

Stworzenie wykresu na podstawie tabeli przestawnej nr 1

```python
worksheet_2 = writer.sheets['Tabela_1']
chart_1 = workbook.add_chart({'type': 'column'})
chart_1.add_series({
    'name':       '=Tabela_1!$B$4',
    'categories': '=Tabela_1!$A$4:$A$247',
    'values':     '=Tabela_1!$B$4:$B$247',
    'fill':       {'color': 'red'},
})
chart_1.set_title ({'name': 'Suma not poszczególnych zawodników biorących udział w zawodach na przestrzeni sezonów 2000/2001 - 2009/2010'})
chart_1.set_y_axis({'name': 'Nota', 'major_gridlines': {'visible': False}})
chart_1.set_x_axis({'name': 'Nazwiska i imiona zawodników'})
chart_1.set_legend({'position': 'none'})
worksheet_2.insert_chart('E2', chart_1)
```

Zapisywanie drugiego raportu do arkusza o nazwie Raport_2

```python
data_2.to_excel(writer, 'Raport_2', index=False)
workbook = writer.book
```

Utworzenie tabeli przestawnej z drugiego raportu

```python
table_2 = pd.pivot_table(data_2, index=['Nazwa skoczni'], values=['Liczba zawodów'],
                              aggfunc=[np.sum], fill_value=0)
table_2
```

Zapis drugiej tabeli przestawnej do arkusza o nazwie Tabela_2

```python
table_2.to_excel(writer, 'Tabela_2')
workbook = writer.book
```

Stworzenie wykresu na podstawie tabeli przestawnej nr 2

```python
worksheet_4 = writer.sheets['Tabela_2']
chart_2 = workbook.add_chart({'type': 'column'})
chart_2.add_series({
    'name':       '=Tabela_2!$B$4',
    'categories': '=Tabela_2!$A$4:$A$32',
    'values':     '=Tabela_2!$B$4:$B$32',
    'fill':       {'color': 'green'},
})
chart_2.set_title ({'name': 'Liczba rozegranych zawodów na przestrzeni sezonów 2000/2001 - 2009/2010 z podziałem na skocznie'})
chart_2.set_y_axis({'name': 'Liczba zawodów', 'major_gridlines': {'visible': False}})
chart_2.set_x_axis({'name': 'Nazwy skoczni'})
chart_2.set_legend({'position': 'none'})
worksheet_4.insert_chart('E2', chart_2)
```

Zapisywanie trzeciego raportu do arkusza o nazwie Raport_3

```python
data_3.to_excel(writer, 'Raport_3', index=False)
workbook = writer.book
```

Utworzenie tabeli przestawnej z trzeciego raportu

```python
table_3 = pd.pivot_table(data_3, index=['Państwo'], values=['Liczba zawodników'],
                              aggfunc=[np.sum], fill_value=0)
table_3
```

Zapis trzeciej tabeli przestawnej do arkusza o nazwie Tabela_3

```python
table_3.to_excel(writer, 'Tabela_3')
workbook = writer.book
```

Stworzenie wykresu na podstawie tabeli przestawnej nr 3

```python
worksheet_6 = writer.sheets['Tabela_3']
chart_3 = workbook.add_chart({'type': 'column'})
chart_3.add_series({
    'name':       '=Tabela_3!$B$4',
    'categories': '=Tabela_3!$A$4:$A$25',
    'values':     '=Tabela_3!$B$4:$B$25',
    'fill':       {'color': 'blue'},
})
chart_3.set_title ({'name': 'Liczba zawodników z poszczególnych państw na przestrzeni sezonów 2000/2001 - 2009/2010'})
chart_3.set_y_axis({'name': 'Liczba zawodników', 'major_gridlines': {'visible': False}})
chart_3.set_x_axis({'name': 'Nazwy państw'})
chart_3.set_legend({'position': 'none'})
worksheet_6.insert_chart('E2', chart_3)
```

Zapisywanie całego programu i wysłanie ostatecznych wyników do pliku Excel

```python
writer.save()
```
